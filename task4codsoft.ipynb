LIBRARIES NEEDED FOR TASK

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
my_dataset= pd.read_csv(r"c:\Users\SWAPNA\Downloads\advertising.csv")
my_dataset.head(10)
```

```python
sns.pairplot(my_dataset)
plt.show()
```

```python
sns.heatmap(my_dataset.corr(),annot=True,cmap='Blues')
plt.show()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(my_dataset[['TV']],my_dataset[['Sales']],test_size=0.3,random_state=0)
```

```python
print(x_train,x_test,y_train,y_test)
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_train,y_train)
```

```python
#now prediction  will take place here
y_pred=model.predict(x_test)
print(y_pred)
```

```python
plt.plot(y_pred)
```
