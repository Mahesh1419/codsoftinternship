```python
class_counting=my_dataset['Class'].value_counts()
plt.bar(class_counting.index,class_counting.values,color=['pink','black'])
plt.xlabel('class')
plt.ylabel('counting')
plt.title('distrubution of fraud vs non fraud transactions')
plt.xticks(ticks=[0,1],labels=['non_fraud','fraud'])
plt.show()
```

```python
corr = my_dataset.corr()
sns.heatmap(corr, annot=False, cmap='viridis', center=0, linewidths=1, linecolor='pink')
plt.title('Correlation Matrix', fontsize=28)
plt.show()
```

```python
from sklearn.model_selection import train_test_split
my_dataset = my_dataset.dropna() 
my_dataset = pd.get_dummies(my_dataset, drop_first=True) 
X = my_dataset.drop('Class', axis=1) 
y = my_dataset['Class'] 
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(random_state=42)
model.fit(X_train, y_train)


```

```python
y_pred=model.predict(X_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
print(classification_report(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
```

```python
sns.heatmap(confusion_matrix(y_test,y_pred),annot=True,cmap='Oranges')
plt.show()
```
